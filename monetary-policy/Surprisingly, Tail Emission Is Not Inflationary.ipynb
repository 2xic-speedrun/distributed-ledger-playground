Choosing to model the supply of blocks as $N(t) = N_0 + kt$ where $k$ is the fixed block reward

### Naive model (no coin loss)

```python
import numpy as np
import matplotlib.pyplot as plt

N_0 = 1000
k = 10
x = np.arange(0, 100)
f = N_0 + x * k
plt.plot(x, f)
```

## Improved model (coin loss rate)

Coin can only be lossed once, and we can model with lambda coin loss constant. You can check out the blogpost to see how the equation is retrived.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import numpy as np
@widgets.interact(min=-10, max=30)
def f(
        lambda_c=widgets.FloatSlider(min=0.0000001, max=1, step=0.0000001, value=0.0000001),
        N_0=widgets.FloatSlider(min=0, max=1000000, step=1, value=1000),
        reward=widgets.FloatSlider(min=0, max=1000000, step=1, value=10)
    ):
    k = reward
    x = np.arange(0, 100)
    f = k/lambda_c + (N_0 - k/lambda_c)*np.exp(-lambda_c*x)
    plt.plot(x, f)
```
